```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("globalterrorismdb_0718dist.csv",engine="python")
data.head()
```

```python
# Renaming the columns 
data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
df=data[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
df.columns
```

```python
df.hist(figsize=(15,10))
```

```python
df.head()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x="Year",data=df)
plt.xticks(rotation="vertical")
plt.title("Terrorist Activity By year")
```

```python
#Number of people killed 
df['Killed'].value_counts().sum()
```

```python
#Number of people wounded
df['Wounded'].value_counts().sum()
```

```python
df['AttackType'].unique()
```

```python
#Attack type
plt.figure(figsize=(15,10))
sns.countplot(x="AttackType",data=df,palette="magma")
plt.xticks(rotation="vertical")
plt.title("Common Attacks")
```

```python
# target type
target= df['Target_type'].value_counts()
target
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=target,y=target.index,data=df,orient='h',palette='magma')
plt.title("Target")
```

```python
#Weapon Type
weapon=df['Weapon_type'].unique()[1:]
weapon
```

```python
df['Weapon_type'].value_counts()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x="Weapon_type",data=df,palette="copper")
plt.xticks(rotation="vertical")
plt.title("Common Weapons")
```

```python
df['Motive'].unique()
```

```python
#Groups
df['Group'].unique()
```

```python
groups=df['Group'].value_counts()[1:10]
groups
```

```python
#Most active groups
plt.figure(figsize=(15,10))
sns.barplot(x=groups,y=groups.index,data=df,orient='h',palette="magma")
plt.title("Most Active Groups")
```

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(20,10))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# most active countries
country=df['Country'].value_counts().nlargest(10)
country
```
